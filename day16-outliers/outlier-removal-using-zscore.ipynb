## Outlier Detection Using Z-score

- Data should be normally distributed or close to normal distribution.
- **Z-score formula:**  
  x’ₙ = (xₙ - μ) / σ  
  where μ is the mean and σ is the standard deviation.
- If x’ₙ is between -3 and 3, the value is **not** an outlier.  
  Anything outside this range is considered an outlier.

### Outlier Treatment  
- Trimming: Remove data points identified as outliers.
- Capping: Set outliers to boundary values.  
  For example, if values above 80 and below 3 are outliers, and data is [2, 85, 90], after capping it becomes [3, 80, 80].


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("placement.csv")
df.head()
```

```python
# see the distribution of the data
plt.figure(figsize=(16,5))
plt.subplot(1,2,1)
sns.kdeplot(df['cgpa'])

plt.subplot(1,2,2)
sns.kdeplot(df['placement_exam_marks'])

plt.show()
```

```python
df['placement_exam_marks'].skew()
```

```python
print("Mean value of cgpa",df['cgpa'].mean())
print("Std value of cgpa",df['cgpa'].std())
print("Min value of cgpa",df['cgpa'].min())
print("Max value of cgpa",df['cgpa'].max())
```

```python
# Finding the boundary values
print("Highest allowed",df['cgpa'].mean() + 3*df['cgpa'].std())
print("Lowest allowed",df['cgpa'].mean() - 3*df['cgpa'].std())
```

```python
# Finding the outliers
df[(df['cgpa'] > 8.80) | (df['cgpa'] < 5.11)]
```

### Trimming

```python
# Trimming

new_df = df[(df['cgpa'] < 8.80) & (df['cgpa'] > 5.11)]
new_df
```

```python
# Approach 2

# Calculating the Zscore

df['cgpa_zscore'] = (df['cgpa'] - df['cgpa'].mean())/df['cgpa'].std()
```

```python
df.head()
```

```python
df[df['cgpa_zscore'] > 3]
```

```python
df[df['cgpa_zscore']< -3]
```

```python
# outliers
df[(df['cgpa_zscore'] > 3) | (df['cgpa_zscore']< -3)]
```

```python
df [df['cgpa_zscore'].between(-3, 3)]
```

### Capping

```python
upper_limit = df['cgpa'].mean() + 3*df['cgpa'].std()
lower_limit = df['cgpa'].mean() - 3*df['cgpa'].std()
```

```python
upper_limit, lower_limit
```

```python
df['cgpa'] = np.where(
    df['cgpa']>upper_limit,
    upper_limit,
    np.where(
        df['cgpa']<lower_limit,
        lower_limit,
        df['cgpa']
    )
)
```
