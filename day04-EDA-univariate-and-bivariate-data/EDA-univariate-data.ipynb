```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("sneakers_streetwear_sales_data.csv")
df.head()
```

## ðŸ§  Working with Categorical Data



### ðŸ“Œ Countplot (Bar Chart)

```python
sns.countplot(x='Product Type', data=df)
```

### ðŸ“Œ PieChart

```python
df['Product Type'].value_counts().plot.pie(autopct='%1.1f%%', figsize=(8, 8))
```

## ðŸ§  Working with Numerical Data


### ðŸ“Œ Histogram

```python
plt.hist(df['Quantity'], bins=5)
```

### ðŸ“Œ Distplot

```python
sns.distplot(df['Quantity'], kde=True, bins=5)
```

### ðŸ“Œ Boxplot

```python
sns.boxplot(df['Quantity'])
```
