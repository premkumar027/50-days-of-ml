```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_regression
```

```python
X, y = make_regression(n_samples=4, n_features=1, n_informative=1, noise=50, random_state=42)
```

```python
plt.scatter(X, y, color='blue')
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
model.coef_, model.intercept_
```

```python
m = -30.39
b = 31.37
```

```python
plt.scatter(X,y)
plt.plot(X,model.predict(X),color='red')
```

```python
y_pred = ((m * X) + 0).reshape(4) # Example prediction with b = 100
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red', linewidth=2)
```

```python
plt.scatter(X,y)
plt.plot(X,model.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='#00a65a',label='b = 0')
plt.legend()
plt.show()
```

```python
loss_slope = -2 * np.sum(y - (m*X.ravel() - b))
print(f"Loss slope: {loss_slope}")
```

```python
# learning rate
lr = 0.01
```

```python
# step size
step_size = lr * loss_slope
print(f"Step size: {step_size}")
```

```python
# calculate new intercept
b = b - step_size
print(f"New intercept: {b}")
```

```python
y_pred1 = (( m * X ) + b).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,model.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

```python
# iteration 2
loss_slope = -2 * np.sum(y - (m*X.ravel() - b))
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred2 = ((m * X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,model.predict(X),color='red',label='OLS')
plt.plot(X,y_pred2,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pred1,color='#A3E4D7',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

```python
b = -100
m 
lr = 0.01

epochs = 100

for i in range(epochs):
  loss_slope = -2 * np.sum(y - m*X.ravel() - b)
  b = b - (lr * loss_slope)

  y_pred = m * X + b

  plt.plot(X,y_pred)

plt.scatter(X,y)
```
